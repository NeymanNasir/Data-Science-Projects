```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
# Given file names
xl_file = 'Efficiency Report A Shift-JANUARY-19.xls'
sheet = 'Sheet1'
repeat_interval = 39 # after how many rows same data format repeat agains
```

```python
# reading raw efficiency data from sheet1
eff_data = pd.read_excel(xl_file,
                        sheet_name = sheet,
                        header = None)
```

```python
eff_data.head(7)
```

```python
# creating a report_date column from the given report dates
report_dates = []
for x in range(0, eff_data.shape[0], 39):
    report_dates.extend([pd.to_datetime(eff_data.iloc[x,3]).date()]*39)

report_dates = report_dates[:eff_data.shape[0]]
```

```python
print(eff_data.shape[0], len(report_dates))
```

```python
# let's take only required columns
required_columns = [0, 1, 2, 3, 4, 6, 14, 16, 19, 22]
required_data = eff_data[required_columns]
```

```python
required_data.columns = ['line', 'buyer', 'style', 'order_no', 'smv', 'day_output', 'npt_hrs', 'total_attended_hrs',\
                        'actual_manpower_ttl', 'total_target']

required_data['date'] = report_dates
```

```python
required_data.head(100)
```

```python
#let's only take required rows and eliminate initial and ending redundant rows for each intervals
```

```python
required_rows = pd.DataFrame(columns = required_data.columns)
cnt = 1
for x in range(0, required_data.shape[0], 39):
    required_rows = pd.concat([required_rows, required_data.iloc[x+5:39*cnt-4, :]], ignore_index = True)
    cnt += 1
```

```python
# Now here we got the required columns and rows for two reporting days
# let's clean existing columns and generate new required columns
```

```python
# column: line
# as this column involves merge cells for each line
# I see we can fill empty values using forwardfill option

required_rows["line"] = required_rows["line"].fillna(method = 'ffill')
```

```python
required_rows.head(10)
```

```python
required_rows.info()
```

```python
# lets calculate column: hours
# hours = Total_Atten.Hrs/actual_manpower_ttl

required_rows["total_attended_hrs"] = pd.to_numeric(required_rows["total_attended_hrs"], errors ='coerce')
required_rows["actual_manpower_ttl"] = pd.to_numeric(required_rows["actual_manpower_ttl"], errors ='coerce')

required_rows["hours"] = (required_rows["total_attended_hrs"]/required_rows["actual_manpower_ttl"]).replace(np.inf, 0)
```

```python
# lost_mins
# Need to convert hours into minutes

required_rows["npt_hrs"] = pd.to_numeric(required_rows["npt_hrs"], errors ='coerce')
required_rows["lost_mins"] = required_rows["npt_hrs"] * 60
```

```python
# available_mins
# Total_Atten.Hrs * 60
required_rows["available_mins"] = required_rows["total_attended_hrs"] * 60
```

```python
# output_mins
# day_output*smv
required_rows["day_output"] = pd.to_numeric(required_rows["day_output"], errors ='coerce')
required_rows["output_mins"] = required_rows["day_output"] * required_rows["smv"]
```

```python
# target_mins
# total_target* smv
required_rows["total_target"] = pd.to_numeric(required_rows["total_target"], errors ='coerce')
required_rows["target_mins"] = required_rows["total_target"] * required_rows["smv"]
```

```python
required_rows
```

```python
# running_days

# Have to count the consecutive days after style change(per report_date, per line)
# strategy: 
# for first day if have values put 1 if null put 0
# from second day if the style value is null then put 0
# if the style value is different than previous value then put 1
# if the style value is same as previuos value than add 1 with the previous day value
```

```python
# our current common row interval for every day is 30
running_days = []
input_date_cal = []

#calculating day1 value
for i in range(30):
    if pd.isnull(required_rows.iloc[i, 2]):
        running_days.append(0)
        input_date_cal.append(np.nan)
    else:
        running_days.append(1)
        input_date_cal.append(required_rows.iloc[i, 10])
        
for i in range(30, required_rows.shape[0], 30):
    for j in range(i, i+30):
        if pd.isnull(required_rows.iloc[j, 2]):
            running_days.append(0)
            input_date_cal.append(np.nan)
        elif required_rows.iloc[j, 2] != required_rows.iloc[j-30, 2]:
            running_days.append(1)
            input_date_cal.append(required_rows.iloc[j, 10])
        else:
            # style matches with previous day
            running_days.append(running_days[j-30]+1)
            input_date_cal.append(required_rows.iloc[j-30, 10])
```

```python
# Adding running_days and input_date_ca columns in the dataframe
required_rows['running_days'] = running_days
required_rows['input_date_cal'] = input_date_cal
```

```python
# Now let's extract only required columns
final = required_rows[['actual_manpower_ttl', 'date', 'buyer', 'line', 'style', 'order_no', 'smv', 'running_days',\
                      'input_date_cal', 'total_target', 'hours', 'lost_mins', 'day_output',\
                      'available_mins', 'output_mins', 'target_mins']]
```
